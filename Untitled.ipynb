## Data Cleaning

### Menghilangkan kolom yang tidak diperlukan

```python
data.drop(['Unnamed: 0',
           'Url_spotify',
           'Uri',
           'Danceability',
           'Energy',
           'Key',
           'Loudness',
           'Speechiness',
           'Acousticness',
           'Instrumentalness',
           'Liveness',
           'Valence',
           'Tempo',
           'Duration_ms'
           'Url_youtube',
           'Title',
           'Channel',
           'Description'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

### Menghilangkan Missing Value

```python
median = data[('Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms')].median()
data['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms'].fillna(median, inplace=True)
```
